**Real World visualization Examples**

**Line Graph**


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas=pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title('Gas Prices over time (in USD)', fontdict={'fontweight':'bold','fontsize':20})
plt.plot(gas.Year, gas.USA, 'b.-',label="US")
plt.plot(gas.Year,gas.Canada, 'r.-',label='Canada',)
plt.plot(gas.Year, gas['South Korea'], 'g.-',label='South Korea')

#Another way to plot may values
# target_countries=['Australia','USA','Canada','South Korea']
# for country in gas:
#     if country in target_countries:
#         plt.plot(gas.Year,gas[country],marker='.', )


plt.xlabel('Year')
plt.ylabel('US Dollars')

#Take data from every 3 years
plt.xticks(gas.Year[::3])
plt.legend()

#to save the figure
plt.savefig('GasPricesFigure.png',dpi=300)
plt.show()
```

**Histograms**

```python
fifa=pd.read_csv('fifa_data.csv')
bins=[40,50,60,70,80,90,100]

#draw a histogram on the 'Overall' column
plt.hist(fifa.Overall, color='#ebb0cd',edgecolor='black')
plt.xticks(bins)

plt.ylabel("Number of layers")
plt.xlabel("Skill Level")
plt.title("Distribution of Player Skills in FIFA 2018", fontdict={'fontweight':'bold'})
plt.show()
```

**Pie Charts**


```python
#how may players of Fifa prefer the left foot?
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['Left','Right']
colors=['#bbd18c','#de795f']
#2F means 2 decimals after the dot
#%% are responsible for the '%' sign
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2F %%')
plt.title('Foot preference of FIFA Players')
plt.show()
                                
```

```python
#fetch the weight from the strings
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
light=fifa.loc[fifa.Weight<125].count()[0]

#[0] is here because we are interested in a count of rows
#and count() returns a Pandas Series (1dimensional labeled array)

light_medium=fifa.loc[(fifa.Weight>=125)&(fifa.Weight<150)].count()[0]
medium=fifa[(fifa.Weight>=150)&(fifa.Weight<175)].count()[0]
medium_heavy=fifa[(fifa.Weight>=175)&(fifa.Weight<200)].count()[0]
heavy=fifa[fifa.Weight>=200].count()[0]
weights=[light,light_medium,medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-200','Over 200']
#explode function breaks the graph apart a bit
#the 'explodiness' corresponds to the position of labels on the previous line
explode=(.5,.2,0,0,.5)
plt.title('Weight distribution of Fifa players (in lbs)')
plt.pie(weights,labels=labels,autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.show()
```

**Box Plot Diagram**

```python
plt.style.use('default')
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall']
manchester=fifa.loc[fifa.Club=='Manchester United']['Overall']

labels = ['Barcelona', 'Madrid','Manchester']

colors = ['#1f77b4', '#f72e14','#ff7f0e'] 

# Creating the boxplots
boxprops = dict(facecolor='white', color='black', linewidth=2)
plt.boxplot([barcelona, madrid,manchester], labels=labels, patch_artist=True, boxprops=boxprops)

# Adding labels and title
plt.title("Professional Soccer Team Comparison")
plt.ylabel('FIFA Overall Rating')

# Setting individual box colors
for box, color in zip(plt.boxplot([barcelona, madrid,manchester], labels=labels, patch_artist=True)['boxes'], colors):
    box.set_facecolor(color)

# Showing the plot
plt.show()
```
